## Gráficas con Python

En este *Notebook* expondremos algunos ejemplos de como elaborar gráficas en Python con la biblioteca **Matplotlib**. A pesar de que hoy en día existen diferentes bibliotecas para graficar, **Matpotlib** se mantiene como una de las más usadas.


---


Lo primero será importar la librería de `numpy`. Para graficar no es obligatorio usar `numpy`, de hecho se puede hacer con listas sin embargo más fácil.

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
```

```python
plt.plot(x)
```

```python
fx = np.sin(x)
```

```python
plt.plot(x,fx)
```

```python
x1 = np.linspace(0,2*np.pi,num=100)
```

```python
fx1 = np.sin(x1)
```

```python
plt.plot(x1,fx1)
```

```python
plt.plot(x1,fx1,'g-')
plt.plot(x,fx,'ro')
```

```python
plt.plot(x1,fx1,'g-')
plt.plot(x,fx,'ro')
plt.title("Función seno")
plt.xlabel("x")
plt.ylabel("f(x)")
```

```python
plt.plot(x1,fx1,'g-',label='sin(x)')
plt.plot(x,fx,'ro',label='sin(x)')
plt.title("Función seno")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(loc=0)
```

```python
plt.plot(x1,fx1,'g-',label='sin(x)')
plt.plot(x,fx,'ro',label='sin(x)')
plt.title("Función seno")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.text(4, 0, 'Hola')
plt.legend(loc=0)
```
